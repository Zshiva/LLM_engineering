```python
import os
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
OLLAMA_BASEE_URL = "http://localhost:11434/v1"
MODEL = "llama3.2"
```

```python
from bs4 import BeautifulSoup
import requests

headers = {
 "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}


def ScrapWebsite(url):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content,"html.parser")
    title = soup.title.string if soup.title else " NO title found"
    if soup.body:
        for irrelevant in soup.body(["script","style","img","input"]):
            irrelevant.decompose()
        text = soup.body.get_text(separator="\n", strip=True)
    else:
        text = ""
    return (title + "\n\n" + text)[:2_000]
        
```

```python
system_prompt = "You are a tech geek. You act as an expert for the new students who wants to learn machine learning and AI in modern world."
user_prompt_prefix = "Prepare the top 20 best ideas on how we can learn machine laerning"

def messages_for(website):
    return[
        {"role":"system", "content":system_prompt},
        {"role":"user", "content":user_prompt_prefix + website}
    ]

def summarise(url):
    website = ScrapWebsite(url)
    ollama_via_openai = OpenAI(base_url = OLLAMA_BASEE_URL, api_key="ollama")
    response = ollama_via_openai.chat.completions.create(
        model = MODEL,
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

```python
def display_summary(url):
    summary = summarise(url)
    display(Markdown(summary))
```

```python
display_summary("https://www.geeksforgeeks.org/machine-learning/machine-learning/")
```
